```python
"""
Syntax dibawah merupakan syntax untuk mengimport/memanggil library library yang akan
digunakan untuk mengolah citra yang ada, library-library yang digunakan antara lain, 
metplotlib.pyplot (plt) dan numpy (np)
"""
import matplotlib.pyplot as plt
import numpy as np
```

```python
"""
Syntax dibawah merupakan inisialisasi dari variabel grayImg, yang mana merupakan
variabel yang digunakan untuk menyimpan citra yang akan diolah dengan menggunakan
method imread dari library matplotlib.pyplot (plt) yang dimana citranya diambil 
dari local komputer dan menampilkannya dengan method imshow dari method yang sama
"""
grayImg = plt.imread('osita-noise.jpeg')
plt.imshow(grayImg, 'gray')
grayImg.shape
```

```python
"""
Syntax dibawah merupakan pendefinisian sebuah function yaitu function padding
yang dimana merupakan function yang digunakan untuk memproses operasi dari
sebuah citra yang dimasukkan kedalam function ini
"""
def padding (img, padValue):
    width = padValue // 2
    result = np.pad(img, (width, width), 'edge')

    return result
```

# MEAN FILTERING

```python
"""
Syntax dibawah merupakan pendefinisian sebuah function yaitu function mean
yang dimana merupakan function yang digunakan untuk memproses operasi mean
filtering sebuah citra yang dimasukkan kedalam function ini
"""
def mean (img, windowSize):
    result = np.zeros(img.shape)
    imgPadded = padding(img, windowSize)

    for i in range (windowSize, imgPadded.shape[0] - windowSize):
        for j in range (windowSize, imgPadded.shape[1] - windowSize):
            sumValue = np.ravel(imgPadded[i - windowSize : i + windowSize + 1, j - windowSize:j + windowSize + 1])
            mean = round(np.sum(sumValue) / (windowSize * windowSize))
            result[i][j] = mean

    return result
```

```python
"""
Syntax dibawah merupakan penginisialisasian dari variabel mean3, 
mean5 dan mean7, yang mana diisi dengan nilai balikan dari function 
yang dipanggil didalamnya yaitu function mean
"""
mean3 = mean(grayImg, 3)
mean5 = mean(grayImg, 5)
mean7 = mean(grayImg, 7)

"""
Syntax dibawah merupakan inisialisasi penggunaan figure dari library matplotlib.pyplot 
(plt) yang digunakan untuk menambahkan subplot agar bisa menampilkan citra secara 
menyamping (side-to-side) antara citra asli dan histogramnya
"""
plt.figure(figsize=(25,25))
plt.subplot(141)
plt.imshow(grayImg, "gray")
plt.title('Original');

"""
Syntax dibawah merupakan inisialisasi penggunaan figure dari library matplotlib.pyplot 
(plt) yang digunakan untuk menambahkan subplot agar bisa menampilkan citra secara 
menyamping (side-to-side) antara citra asli dan histogramnya
"""
plt.subplot(142)
plt.imshow(mean3, "gray")
plt.title('Mean 3x3');

"""
Syntax dibawah merupakan inisialisasi penggunaan figure dari library matplotlib.pyplot 
(plt) yang digunakan untuk menambahkan subplot agar bisa menampilkan citra secara 
menyamping (side-to-side) antara citra asli dan histogramnya
"""
plt.subplot(143)
plt.imshow(mean5, "gray")
plt.title('Mean 5x5');

"""
Syntax dibawah merupakan inisialisasi penggunaan figure dari library matplotlib.pyplot 
(plt) yang digunakan untuk menambahkan subplot agar bisa menampilkan citra secara 
menyamping (side-to-side) antara citra asli dan histogramnya
"""
plt.subplot(144)
plt.imshow(mean7, "gray")
plt.title('Mean 7x7');

plt.show()
```

# MEDIAN FILTERING

```python
"""
Syntax dibawah merupakan pendefinisian sebuah function yaitu function median
yang dimana merupakan function yang digunakan untuk memproses operasi median
filtering sebuah citra yang dimasukkan kedalam function ini
"""
def median(img, windowSize):
    result = np.zeros(img.shape)
    imgPadded = np.pad(img, (windowSize, windowSize), 'edge')
    for i in range (windowSize, imgPadded.shape[0] - windowSize):
        for j in range (windowSize, imgPadded.shape[1] - windowSize): 
            temp = np.ravel(imgPadded[i - windowSize : i + windowSize + 1, j - windowSize : j + windowSize + 1])
            temp.sort()
            result[i - windowSize, j - windowSize] = temp[int(len(temp) / 2)]

    return result
```

```python
"""
Syntax dibawah merupakan penginisialisasian dari variabel med3, 
med5 dan med7, yang mana diisi dengan nilai balikan dari function 
yang dipanggil didalamnya yaitu function median
"""
med3 = median(grayImg, 3)
med5 = median(grayImg, 5)
med7 = median(grayImg, 7)

"""
Syntax dibawah merupakan inisialisasi penggunaan figure dari library matplotlib.pyplot 
(plt) yang digunakan untuk menambahkan subplot agar bisa menampilkan citra secara 
menyamping (side-to-side) antara citra asli dan histogramnya
"""
plt.figure(figsize=(25,25))
plt.subplot(141)
plt.imshow(grayImg, "gray")
plt.title('Original');

"""
Syntax dibawah merupakan inisialisasi penggunaan figure dari library matplotlib.pyplot 
(plt) yang digunakan untuk menambahkan subplot agar bisa menampilkan citra secara 
menyamping (side-to-side) antara citra asli dan histogramnya
"""
plt.subplot(142)
plt.imshow(med3, "gray")
plt.title('Median 3x3');

"""
Syntax dibawah merupakan inisialisasi penggunaan figure dari library matplotlib.pyplot 
(plt) yang digunakan untuk menambahkan subplot agar bisa menampilkan citra secara 
menyamping (side-to-side) antara citra asli dan histogramnya
"""
plt.subplot(143)
plt.imshow(med5, "gray")
plt.title('Median 5x5');

"""
Syntax dibawah merupakan inisialisasi penggunaan figure dari library matplotlib.pyplot 
(plt) yang digunakan untuk menambahkan subplot agar bisa menampilkan citra secara 
menyamping (side-to-side) antara citra asli dan histogramnya
"""
plt.subplot(144)
plt.imshow(med7, "gray")
plt.title('Median 7');

plt.show()
```

# MODUS FILTERING

```python
"""
Syntax dibawah merupakan pendefinisian sebuah function yaitu function findModus
yang dimana merupakan function yang digunakan untuk memproses operasi modus
filtering sebuah citra yang dimasukkan kedalam function ini
"""
def findModus(array):
    temp = np.zeros(int(np.max(array))+1)
    for i in array:
        temp[int(i)]+=1
    max = np.max(temp)
    for i in range(len(temp)):
        if(max == temp[i]): 
            return i

"""
Syntax dibawah merupakan pendefinisian sebuah function yaitu function normalisasi
yang dimana merupakan function yang digunakan untuk memproses operasi normalisasi
sebuah citra yang dimasukkan kedalam function ini
"""
def normalisasi(img):
  min = np.min(img)
  max = np.max(img)

  return np.round((img-min)*255/(max-min))

"""
Syntax dibawah merupakan pendefinisian sebuah function yaitu function modus
yang dimana merupakan function yang digunakan untuk memproses operasi modus
filtering sebuah citra yang dimasukkan kedalam function ini
"""
def modus(img, windowSize):
    result = np.zeros(img.shape)
    imgPadded = np.pad(img, (windowSize, windowSize), 'symmetric')
    for i in range(windowSize, imgPadded.shape[0] - windowSize):
        for j in range(windowSize, imgPadded.shape[1] - windowSize): 
            temp = np.ravel(imgPadded[i - windowSize : i + windowSize + 1, j - windowSize : j + windowSize + 1])
            temp.sort()
            result[i - windowSize, j - windowSize] = findModus(temp)

    return normalisasi(result)
```

```python
"""
Syntax dibawah merupakan penginisialisasian dari variabel mod3, 
mod5 dan mod7, yang mana diisi dengan nilai balikan dari function 
yang dipanggil didalamnya yaitu function modus
"""
mod3 = modus(grayImg, 3)
mod5 = modus(grayImg, 5)
mod7 = modus(grayImg, 7)

"""
Syntax dibawah merupakan inisialisasi penggunaan figure dari library matplotlib.pyplot 
(plt) yang digunakan untuk menambahkan subplot agar bisa menampilkan citra secara 
menyamping (side-to-side) antara citra asli dan histogramnya
"""
plt.figure(figsize=(20,20))
plt.subplot(241)
plt.imshow(grayImg, "gray")
plt.title('Original');

"""
Syntax dibawah merupakan inisialisasi penggunaan figure dari library matplotlib.pyplot 
(plt) yang digunakan untuk menambahkan subplot agar bisa menampilkan citra secara 
menyamping (side-to-side) antara citra asli dan histogramnya
"""
plt.subplot(242)
plt.imshow(mod3, "gray")
plt.title('Modus 3x3');

"""
Syntax dibawah merupakan inisialisasi penggunaan figure dari library matplotlib.pyplot 
(plt) yang digunakan untuk menambahkan subplot agar bisa menampilkan citra secara 
menyamping (side-to-side) antara citra asli dan histogramnya
"""
plt.subplot(243)
plt.imshow(mod5, "gray")
plt.title('Modus 5x5');

"""
Syntax dibawah merupakan inisialisasi penggunaan figure dari library matplotlib.pyplot 
(plt) yang digunakan untuk menambahkan subplot agar bisa menampilkan citra secara 
menyamping (side-to-side) antara citra asli dan histogramnya
"""
plt.subplot(244)
plt.imshow(mod7, "gray")
plt.title('Modus 7x7');

plt.show()
```
